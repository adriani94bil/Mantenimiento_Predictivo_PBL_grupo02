```python
import pandas as pd
import numpy as np

horizontal_05mm12 = pd.read_csv('../bearing_fault/horizontal-misalignment/0.5mm/12.288.csv')

# Insertar una fila vacía al principio
horizontal_05mm12.loc[-1] = [np.nan] * len(horizontal_05mm12.columns)  # Crear una fila vacía
horizontal_05mm12.index = horizontal_05mm12.index + 1  # Desplazar los índices hacia abajo
horizontal_05mm12 = horizontal_05mm12.sort_index()  # Ordena el índice para que la fila vacía quede en la parte superior

# Asegurar que la primera fila de datos sea la fila correcta con los valores -0.49977
horizontal_05mm12.iloc[0] = [-0.49977, 0.97159, 0.086419, -0.01759, 0.64014, 0.19493, 0.35288, 0.021614]

# Asignar nombres de columnas si es necesario
column_names = ['S1', 'S2', 'S3', 'S4', 'S5', 'S6', 'S7', 'S8']
horizontal_05mm12.columns = column_names  # Asignar nombres de columnas

horizontal_05mm12


```

```python
# Número de muestras por segmento (0.01 segundos APROX)
segment_length = 500

# Número total de segmentos completos que se pueden hacer
num_segments = len(horizontal_05mm12) // segment_length  # Número de bloques completos


segments = [horizontal_05mm12.iloc[i * segment_length: (i + 1) * segment_length] for i in range(num_segments)]
```

# Atributos estadisticos

```python
df = segments[1]

# Crear las estadísticas para cada columna de manera automática
stats = {}

# Iterar por cada columna y calcular las estadísticas
for column in df.columns:
    stats[f'{column}_mean'] = df[column].mean()
    stats[f'{column}_median'] = df[column].median()
    stats[f'{column}_std'] = df[column].std()
    stats[f'{column}_var'] = df[column].var()
    stats[f'{column}_cv'] = df[column].std() / df[column].mean()
    stats[f'{column}_min'] = df[column].min()
    stats[f'{column}_max'] = df[column].max()
    stats[f'{column}_IQR'] = df[column].quantile(0.75) - df[column].quantile(0.25)

# Convertir el diccionario en un DataFrame de una sola fila
stats_df = pd.DataFrame(stats, index=[0])

stats_df
```

# Funcion custom column

```python
import pandas as pd

def add_custom_columns(df, col_names, values):
    for col_name, value in zip(col_names, values):
        df[col_name] = value
        
    return df


custom_column_names = ['Hz', 'medida', 'Tipo']
custom_values = [
    [12.288],  
    ['0.5mm'],   
    ['horizontal_misalignment']  
]

updated_stats_df = add_custom_columns(stats_df, custom_column_names, custom_values)

updated_stats_df
```
