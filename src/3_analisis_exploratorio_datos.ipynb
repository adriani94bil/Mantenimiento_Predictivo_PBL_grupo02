## 3.6 Matrices de correlación

```python
import matplotlib.pyplot as plt
import seaborn as sns

ignore_cols = ['Tipo', 'medida', 'Hz']

fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(16, 24))
axes = axes.flatten()

# Iteramos de S1_ a S8_
for i in range(1, 9):
    prefix = f'S{i}_'
    # Filtramos columnas que empiecen con S{i}_ y no estén en ignore_cols
    col_subset = [col for col in df.columns 
                  if col.startswith(prefix) and col not in ignore_cols]

    ax = axes[i - 1]

    if not col_subset:
        ax.set_visible(False)
        continue

    sub_df = df[col_subset]

    corr_matrix = sub_df.corr()

    sns.heatmap(
        corr_matrix, 
        annot=True, 
        cmap='coolwarm', 
        center=0, 
        ax=ax
    )
    ax.set_title(f'Correlation Matrix for {prefix}')

    # --------------------------------------------
    # DETECTAR E IMPRIMIR CORRELACIONES > 0.85
    # --------------------------------------------
    threshold = 0.85
    # Pasar a formato "largo" y tomar valor absoluto para filtrar
    stacked_corr = (
        corr_matrix
        .abs()
        .stack()
        .reset_index()
    )
    stacked_corr.columns = ['Feature1', 'Feature2', 'Correlation']
    
    # Evitar duplicados y diagonal
    stacked_corr = stacked_corr[stacked_corr['Feature1'] < stacked_corr['Feature2']]

    # Filtrar correlaciones >= threshold
    high_corr = stacked_corr[stacked_corr['Correlation'] >= threshold]
    if not high_corr.empty:
        print(f"\n--- Correlaciones >= {threshold} para {prefix} ---")
        for _, row in high_corr.iterrows():
            f1, f2 = row['Feature1'], row['Feature2']
            # Tomamos la correlación real (puede ser negativa o positiva)
            real_corr = corr_matrix.loc[f1, f2]
            print(f"* {f1} vs {f2}: {real_corr:.2f}")
    else:
        print(f"\nNo hay correlaciones >= {threshold} para {prefix}.")

fig.tight_layout()
plt.show()
```
