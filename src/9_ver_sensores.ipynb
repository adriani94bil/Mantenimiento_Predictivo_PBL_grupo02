# 9. Ver sensores

El objetivo de este notebook es ver que sensor pertenece a que serie temporal con el fin de de ser capaces de identificar cada uno de estos.

Sensores:

* **S1:** tachometer
* **S2:** acc_under_axial
* **S3:** acc_under_radiale
* **S4:** acc_under_tangencial
* **S5:** acc_over_axial
* **S6:** acc_over_radiale
* **S7:** acc_over_tangencial
* **S8:** microphone

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../bearing_fault/normal/12.288.csv')
df
```

```python
if 'time_column_name' in df.columns:  
    df['time_column_name'] = pd.to_datetime(df['time_column_name'])  
    df.set_index('time_column_name', inplace=True)

for column in df.columns:
    plt.figure(figsize=(10, 5))
    plt.plot(df.index, df[column], label=column)
    plt.title(f"Serie temporal de {column}")
    plt.xlabel("Índice")
    plt.ylabel(column)
    plt.legend()
    plt.grid()
    plt.show()
```
